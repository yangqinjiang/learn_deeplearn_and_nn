```python
import tensorflow as tf
# 
value2d = tf.constant(
    [
        [5,1,4,2],
        [3,9,5,7]
    ],
    tf.float32)

#
session=tf.Session()
# 0 
sum0 = tf.reduce_sum(value2d,axis=0)
print("0:",session.run(sum0))
# 1 
sum1 = tf.reduce_sum(value2d,axis=1)
print("1:",session.run(sum1))

#(0,1) 
sum01 = tf.reduce_sum(value2d,axis=(0,1))
print("(0,1) :",session.run(sum01))
```
