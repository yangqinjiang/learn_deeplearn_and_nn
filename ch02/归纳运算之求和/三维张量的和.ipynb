```python
import tensorflow as tf
#
value3d = tf.constant(
[
    [[2,5],[3,3],[8,2]],
    [[6,1],[1,2],[5,4]],
    [[7,9],[2,-3],[-1,3]]
],tf.float32)

#
session=tf.Session()
#0
sum0=tf.reduce_sum(value3d,axis=0)
print('0:',session.run(sum0))

#1
sum1=tf.reduce_sum(value3d,axis=1)
print('1',session.run(sum1))

#(0,1)
sum01=tf.reduce_sum(value3d,axis=(0,1))
print('(0,1)',session.run(sum01))

#(0,1,2)
sum012=tf.reduce_sum(value3d,axis=(0,1,2))
print('(0,1,2)',session.run(sum012))
```
